# Sorting:

```python
unsorted = np.array([3,1,6,2,6,1,8,5])
print("sorted:", np.sort(unsorted))

arr_2d_unsorted = np.array([[9,1],
                            [7,9],
                            [1,7]])
print("sorted 2d array:", np.sort(arr_2d_unsorted ,axis=0)) # top to bottom 
print("sorted 2d array:", np.sort(arr_2d_unsorted ,axis=1)) # right and left
```

# Filter:

```python
numbers = np.array([2,6,21,5,32,9])
even_numbers = numbers[numbers % 2 == 0]
print("even number: ", even_numbers)
```

# Fancy Indexing and np.where()

```python
indices = [0,2,4] # positons/index
print(numbers[indices])

where_result = np.where(numbers > 5)
print("NP Where: ", numbers[where_result])

```

# Combine the 2 Arrays:

```python
import numpy as np
arr1 = np.array([3,6,1])
arr2 = np.array([5,6,1])

combined_array = np.concatenate((arr1, arr2))
print("combined array: ", combined_array)
```

# Array compatibilty , Same Shape:

```python
a = np.array([3,7,12])
b = np.array([5,8,2,])
c = np.array([7,2,6])

print("shape of a: ", a.shape)
print("shape of b: ", b.shape)
print("shape of c: ", c.shape)

print("Compatibilty shapes: ", a.shape == b.shape)
```

# Add new row and column in array:

```python
original_array = np.array([[2,5], [4,8]])
print("original array: ", original_array)
new_row = np.array([[7,3]])
nr = np.vstack((original_array , new_row)) #vstack for row (vertical)
print("original_array with new row: ", nr )

new_col = np.array([[8], [1]])
nc = np.hstack((original_array, new_col)) #hstack for column(horizontal)
print("original array with new column: ", nc)
```
