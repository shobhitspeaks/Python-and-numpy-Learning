# Advance operation with Business Examples:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure: [restaurant_id , 2021, 2022 ,2023 , 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #paradise Biryani
    [2, 120000, 140000, 160000, 190000], #Beijing Bites
    [3, 200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000], #Burger Point
    [5, 160000, 185000, 205000, 230000], #Chai Point
])
print("==== Zomato Sales Analysis ===")
print("shape of data: ", sales_data.shape)
print("First 3 Sales data: ", sales_data[0:3])
```

```python
# axis = 0 is for vertical wise means columns
# axis = 1 is for horizontal wise means rows
```

```python
# Total sales per year:
yearly_total = np.sum(sales_data [:,1:], axis=0) # 1 is for ignore restaurant_id
print("2021 total sales: ", yearly_total[0])
print("2022 total sales: ", yearly_total[1])
print("2023 total sales: ", yearly_total[2])
print("2024 total sales: ", yearly_total[3])

# minimum sales per restaurant:
min_sales = np.min(sales_data[:,1:], axis = 1)
print("minimum sales per restaurant: ", min_sales)

#Maximum sales per year:
max_sales = np.max(sales_data [:,1:], axis = 0)
print("maximum sales: ", max_sales)


#Average sales per restaurant:
avg_sales = np.mean(sales_data [:,1:], axis=1)
print("average sales: ", avg_sales)

#Cumulative sales:
cumsum = np.cumsum(sales_data [:,1:], axis = 1)
print("cumulative sales: ", cumsum)
```

```python
#cumulative sales graph by matplotlib:
plt.Figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

# Addition, Multiplication,  angle and Dot Product of Vectors:

```python
vector1 = np.array([1,3,4,2,6])
vector2 = np.array([4,8,2,8,2])

print("Addition of Vectors: ", vector1 + vector2)
print("Mulitiplication of Vectors: ", vector1 * vector2)
print("Dot Product of Vectors: ", np.dot(vector1, vector2))

# Angle between vectors:
angle = np.arccos(np.dot(vector1,vector2) / (np.linalg.norm(vector1)* np.linalg.norm(vector2))) 
print("angle of vector: " ,angle)
```
