# List Vs Array:

```python
py_list = [23,56,21]
print("multiplication of list: ", py_list * 2)

array = np.array([23,56,21]) #element wise multiply
print("multiplication of array: ", array * 2)

import time

start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List Operation time is: ", time.time()-start)

start = time.time()
array = np.arange(1000000)*2
print("\n Array Operaton time is: ", time.time()-start)
```

# Creating arrays from scratch

```python
import numpy as np
zeros = np.zeros((3,3))
print("zeros array: \n", zeros)

ones = np.ones((3,3))
print("ones array: \n", ones)

full = np.full((3,3), 20) #constant value
print("Full array: \n", full)

random  = np.random.random((2,3)) #random numbers
print("random array: \n", random)

sequence = np.arange(0,21,2)
print("squence array: \n", sequence)

```

# Vector , Matrice and Tensor:

```python
import numpy as np
vector = np.array([2,6,7])
print("vector:",vector)

Matrix = np.array([[3,6,1],[6,9,12]])    #2 dimesions
print("Matrice: ",Matrix)

Tensor = np.array([[[3,2],[7,1],[6,3],[3,2]]]) # more than 2 dimension
print("tensor: ", Tensor)
```

# Array Properties:

```python
array = np.array([[2,3,5],[6,9,1]])
print("shape:", array.shape) # array's shape
print("size:", array.size) # array's size
print("type:", array.dtype) # array's type
print("dimension:", array.ndim) # array's dimensions

array = np.array([[2,3,5],[6,9,1.44]]) #float value
print("type:", array.dtype) # array's type

array = np.array([[2,3,5],[6,9,"1.44"]]) #String value
print("type:", array.dtype) # array's type

```

# Array Reshaping:

```python
import numpy as np
array = np.arange(10)
print("Original Array:", array)

Reshape = array.reshape((2,5))
print("reshaped array: ", Reshape)

flatttend = Reshape.flatten() # it provides copy
print("flattend array: ", flatttend)

raveled = Reshape.ravel() # it returns original array, not a copy
print("raveled array: ", raveled) 

transpose = Reshape.T
print("transposed array: ", transpose)

```
